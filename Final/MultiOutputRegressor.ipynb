```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Crop_recommendation.csv')
```

Visualizing the relations between the dataset features and label

```python
import seaborn as sns

sns.pairplot(data[['N','P','K','label']])
```

The crop that needs to be predicted depends all the features in the dataset as Nitrogen,Phosphorous,Potassium,Rainfall,Temperature,pH all are independent features

```python
unique_crops = data['label'].unique()
print(f'The unique crops that can be predicted are : {unique_crops}')
```

```python
data.describe()
```

```python
max_ph_value = data['ph'].max()
min_ph_value = data['ph'].min()

print(f"The overall ph value of the soil samples is in between {min_ph_value} : {max_ph_value}")
```

```python
max_rainfall = data['rainfall'].max()
min_rainfall = data['rainfall'].min()

print(f"The rainfall to be expected is in between : {min_rainfall} : {max_rainfall}")
```

```python
plt.plot(data.describe())
plt.title('Understanding the trend of the dataset')
plt.figure(figsize=(20,10))
plt.show()
```
