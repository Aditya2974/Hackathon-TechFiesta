```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

def plot_gantt_chart(farmers_data):
    fig, ax = plt.subplots(figsize=(10, len(farmers_data)*0.5))

    for i, (farmer, crops) in enumerate(farmers_data.items()):
        for crop, start_date, end_date in crops:
            start = datetime.strptime(start_date, "%Y-%m-%d")
            duration = (datetime.strptime(end_date, "%Y-%m-%d") - start).days
            ax.barh(i, duration, left=start, height=0.5, align='center', color='blue', alpha=0.6)
            ax.text(start + duration/2, i, crop, ha='center', va='center', color='black')

    ax.set_yticks(np.arange(len(farmers_data)))
    ax.set_yticklabels(list(farmers_data.keys()))
    ax.invert_yaxis()  # Invert y-axis to have the farmers listed from top to bottom
    ax.set_xlabel('Time')
    ax.set_title('Gantt Chart')

    plt.show()

# Example usage:
farmers_data = {
    'Farmer 1': [('Wheat', '2023-01-01', '2023-06-01'), ('Corn', '2023-07-01', '2023-12-01')],
    'Farmer 2': [('Soybean', '2023-02-01', '2023-05-01'), ('Barley', '2023-06-01', '2023-10-01')],
    'Farmer 3': [('Rice', '2023-03-01', '2023-08-01')],
    'Farmer 4': [('Potato', '2023-04-01', '2023-09-01')],
    'Farmer 5': [('Cotton', '2023-01-01', '2023-12-01')]
}

plot_gantt_chart(farmers_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta
from matplotlib.dates import YearLocator, DateFormatter

def plot_gantt_chart(farmer_data):
    fig, ax = plt.subplots(figsize=(18, 8))

    for i, (crop, start_date, end_date) in enumerate(farmer_data):
        start = datetime.strptime(start_date, "%Y-%m-%d")
        end = datetime.strptime(end_date, "%Y-%m-%d")
        duration = (end - start).days
        ax.barh(i, duration, left=start, height=0.8, align='center', color='blue', alpha=0.6)
        ax.text(start + timedelta(days=duration / 2), i, crop, ha='center', va='center', color='black')

    ax.set_yticks(np.arange(len(farmer_data)))
    ax.set_yticklabels([crop for crop, _, _ in farmer_data])
    ax.invert_yaxis()  # Invert y-axis to have the crops listed from top to bottom
    ax.set_xlabel('Time')
    ax.set_title('Gantt Chart')

    # Set x-axis ticks to show only the year
    ax.xaxis.set_major_locator(YearLocator())
    ax.xaxis.set_major_formatter(DateFormatter('%y'))

    # Adjust x-axis limits for better visibility of ticks
    ax.set_xlim(left=min([datetime.strptime(start, "%Y-%m-%d") for _, start, _ in farmer_data]),
                right=max([datetime.strptime(end, "%Y-%m-%d") for _, _, end in farmer_data]) + timedelta(days=365))

    plt.show()

# Example usage:
farmer_data = [
    ('Wheat', '2019-03-15', '2019-05-01'),
    ('Corn', '2019-07-01', '2019-09-01'),
    ('Soybean', '2020-02-15', '2020-05-15'),
    ('Barley', '2020-08-01', '2020-10-01'),
    ('Rice', '2021-03-01', '2021-07-01'),
    ('Potato', '2021-05-01', '2021-08-01'),
    ('Maize', '2022-04-01', '2022-07-01'),
    ('Sorghum', '2022-07-01', '2022-10-01'),
    ('Sunflower', '2023-03-01', '2023-06-01'),
    ('Cotton', '2023-06-01', '2023-09-01'),
    ('Peanut', '2024-02-01', '2024-06-01'),
    ('Sugarcane', '2024-05-01', '2024-08-01'),
]

plot_gantt_chart(farmer_data)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

def plot_gantt_chart(tasks):
    fig, ax = plt.subplots(figsize=(14, len(tasks)*1.5))

    for i, task in enumerate(tasks):
        start_date = datetime.strptime(task['start'], '%Y-%m-%d')
        end_date = datetime.strptime(task['end'], '%Y-%m-%d')
        duration = end_date - start_date

        ax.barh(i, width=duration.days, left=start_date, height=0.5, align='center', color='skyblue', alpha=0.7)
        ax.text(start_date + duration/2, i, task['task_name'], ha='center', va='center', color='black', fontsize=10)

    ax.set_yticks(range(len(tasks)))
    ax.set_yticklabels([task['task_name'] for task in tasks])
    ax.xaxis_date()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

    ax.set_xlabel('Timeline')
    ax.set_ylabel('Tasks')

    plt.tight_layout()
    plt.show()

# Example tasks for sugarcane cultivation
sugarcane_tasks = [
    {'task_name': 'Land preparation', 'start': '2024-04-01', 'end': '2024-04-10'},
    {'task_name': 'Planting', 'start': '2024-04-11', 'end': '2024-05-01'},
    {'task_name': 'Irrigation and Fertilization', 'start': '2024-05-02', 'end': '2024-06-01'},
    {'task_name': 'Weeding', 'start': '2024-06-02', 'end': '2024-07-01'},
    {'task_name': 'Harvesting', 'start': '2024-08-01', 'end': '2024-09-01'},
]

plot_gantt_chart(sugarcane_tasks)
```
